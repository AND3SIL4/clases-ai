- Continuas: toman el valor de una recta numerica que tiende hacia el infinito. Deben ser cuantificables. Si se pueden operar

```python
# Manipulacion de matrices
import pandas as pd

# Operaciones aritmeticas
import numpy as np
```

El set de datos de **nombres** contiene los registros de la cantidad de personas con un nombre especifico, nacidas por año en cada estado de Estados Unidos, desde 1910 hasta 2017 (casi 6 millones de registros)

Pandas permite leer diversos tipos de archivo, como csv, json, xlsx, etc. usualmente los datos a utilizar en inteligencia artificial se almacenan como csv

```python
# Leer el archivo nombre.csv y convertirlo a df (read_fileType)
df_base = pd.read_csv('nombres.csv')

# Ver los primeros registros del df
df_base.head(10)
```

```python
# Mostrar informacion sobre el df creado (info)
df_base.info()
```

# Filtros en pandas

El primer registro muestra que en el estdo de MO nacienron 611 mujeres llamadas Mary, durante el año 1910. Ante esto podriamos plantearnos las siguientes interrogantes

* ¿Hay hombres llamados Mary?
* ¿Cual es la cantidad de hombres llamados Mary por cada estado?
* ¿Cual es la cantidad promedio de hombres llamados Mary en cada estado?
* ¿Cuales son los tres estados donde hay mas hombres llamados Mary?

```python
# Obtener todos los registros de hombres llamados Mary
# Almacenar el resultado del filtro en una nueva variable

df_mary = df_base[
    (df_base['sex'] == 'M') &
    (df_base['name'] == 'Mary')
]

# Consultamos los primeros 5 resgistros
df_mary.head()
```

```python
# Generear un nuevo df con las columnas personalizadas
nuevo_df = df_mary[['state', 'quantity']]
nuevo_df.head()
```

```python
# Saber cunados hombre Mary han habido en total (sum)
nuevo_df[['quantity']].sum()
```

```python
# Agrupar por estado utilizando la funcion (groupby) y sumar (sum)
estados_df = nuevo_df.groupby('state').agg('sum')
estados_df.head()
```

```python
# Agrupar por estado usando la funcion (groupby), mirar el promedio (mean)
estados_df = nuevo_df.groupby('state').agg('sum')
estados_df.head()
```

```python
# Ordenar el set de datos por la mayor cantidad
estados_df.sort_values('quantity', ascending=False).head(10)
```

```python
# Acceder a los registros mediante el nuevo indice
estados_df.loc['NY']
```

```python
# Resetear el indice del df
estados_df.reset_index(inplace=True)
estados_df.head(10)
```

Se pueden exportar los set de datos modificados en diferentes extenciones de archivos

```python
estados_df.to_excel('estados_mary.xlsx', index=None)
```

## Reponder las siguientes preguntas utilizando estadistica y pandas

* ¿Cual es el año con la mayor cantidad de Maries en total?
* ¿Cual es el estado que tiene la mayor cantidad de Maries en total?
* ¿Cuantos Maries nacieron entre 2000 y 2010 en total?
* ¿En que año hubo mas Maries en el estado de California?

```python
# Año con mayor cantidad de Maries en total

new_data = df_mary[['year', 'quantity']]
new_data.sort_values('quantity', ascending=False).head(1)
```

```python
# Estado con mayor cantidad de Maries

new_data = df_mary[['state', 'quantity']]
new_data.sort_values('quantity', ascending=False).head(1)
```

```python
# Cuantas Maries nacieron entre 2000 y 2010
marie_df = df_base[
    (df_base['name'] == 'Mary') &
    (df_base['year'] >= 2000) &
    (df_base['year'] <= 2010)
]
new_frame = marie_df [['quantity']]

new_frame.sort_values('quantity', ascending=False).agg('sum')
```

```python
 Año que hubo mas Maries en California
marie_state_df = df_base[
    (df_base['name'] == 'Mary') &
    (df_base['state'] == 'CA')
]

new_df = marie_state_df[['state', 'quantity']]

new_df.sort_values('quantity', ascending=False).head(1)
```
